```python
def parse_data(num):
    
    data_file = "./data/articles_" + str(num_articles) + ".train"

    # data_truth = "./data/articles_" + str(num_articles) + ".truth"

    f = open(data_file,'r')

    for line in f:

        words = line.split(" ")

        # the article id of each line
        id = words[0]

        article_id.append(id)

        # the article of each line
        text = line.strip(words[0])

        del words[0]

        # remove all punctuation:     
        remove_punct = str.maketrans(dict.fromkeys(string.punctuation))
        text = text.translate(remove_punct)

        # change all letters to lowercase and remove space:
        text = text.lower().replace(" ","")

        # remove all whitespace:
        text = text.strip()

        # creat tuple(id,string) for each line:
        article_tuple = (id, text)

        # Maintain a list of each article_tuple:
        articles_all.append(article_tuple)

    f.close()
    return
```

```python
def shingle_document(text,k):

    shingle_crc = set()  #之前把这行放到了最外面，结果id_shingle_set{}输出的结果一直不是想要的，根据例子放到这里解决了问题,因为每次要清零。

    for index in range(0,len(text) - int(k) + 1): #text, text_id，在后面的主体程序里定义了

        shingle = text[index: index + int(k)]

        # Hash the shingle to a 32-bit integer, encode it before sending e.g. message.encode('utf-8')
        crc = binascii.crc32(shingle.encode('utf-8')) & 0xffffffff

        shingle_crc.add(crc)
        
    id_shingle_set[text_id] = shingle_crc

    return id_shingle_set
```

```python
def jaccard(a,b):
    
    intersection_size = len(a.intersection(b))
    union_size = len(a.union(b))
    Jsim = intersection_size / union_size
    
    #for test:
    #print("intersection_size:",intersection_size)
    #print("union_size:",union_size)
    #print("Jsim in function: ",Jsim,'\n')
    
    return Jsim
```
